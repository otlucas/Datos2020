```python
train = pd.read_csv("data/train.csv",encoding = "ISO-8859-1")
train.shape
```

```python
train.head(20)
```

```python
train['text'] = train['text'].replace(r'http\S+', '', regex=True)
targets = train.groupby('text').target.transform('mean').to_frame()
targets['target'] = targets['target'].round(0).astype('int64')
train['target'] = targets['target']
train = train.drop_duplicates(subset='text')
train.shape
```

```python
train.to_csv('data/train_limpio.csv', index=False)
```
